```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('a1images/daisy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]

mask = np.zeros((height, width), np.uint8)
bgd_model = np.zeros((1, 65), np.float64)
fgd_model = np.zeros((1, 65), np.float64)

rect = (50, 20, width-57, height-40)

cv2.grabCut(img, mask, rect, bgd_model, fgd_model, 5, cv2.GC_INIT_WITH_RECT)
```

```python
final_mask = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')

foreground = img * final_mask[:, :, np.newaxis]
background = img * (1 - final_mask[:, :, np.newaxis])
```

```python
plt.figure(figsize=(12, 20))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(mask, cmap='gray')
plt.title('Mask')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(foreground)
plt.title('Foreground')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(background)
plt.title('Background')
plt.axis('off')
plt.show()
```

```python
blur_strength = 25 
blurred_img = cv2.GaussianBlur(img, (blur_strength, blur_strength), 0)
enhanced_img = np.where(final_mask[:, :, np.newaxis] == 1, img, blurred_img)

blur_strength2 = 65
blurred_img2 = cv2.GaussianBlur(img, (blur_strength2, blur_strength2), 0)
enhanced_img2 = np.where(final_mask[:, :, np.newaxis] == 1, img, blurred_img2)

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image', fontsize=14)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(enhanced_img)
plt.title('Blurred Background strength 25', fontsize=14)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(enhanced_img2)
plt.title('Blurred Background strength 65', fontsize=14)
plt.axis('off')

plt.tight_layout()
plt.show()
```
