```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'a1images/emma.jpg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off') 
plt.show()
```

```python
xp = [0, 50, 150, 255]
fp = [0, 100, 150, 255]

x = np.arange(256)
lookup_table = np.interp(x, xp, fp).astype(np.uint8)

transformed_img = cv2.LUT(img, lookup_table)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(transformed_img, cmap="gray")
axes[1].set_title("Transformed Image")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
