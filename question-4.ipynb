```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('a1images/spider.png')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_hsvFloat=cv2.cvtColor(img,cv2.COLOR_BGR2HSV).astype(np.float32)
plt.imshow(img_hsv)
plt.axis('off')
```

```python
H, S, V = cv2.split(img_hsvFloat)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(H)
plt.title('H')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(S)
plt.title('S')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(V)
plt.title('V')
plt.axis('off')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(H,cmap='gray')
plt.title('H')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(S,cmap='gray')
plt.title('S')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(V,cmap='gray')
plt.title('V')
plt.axis('off')
```

```python
def vibrance_transform(x, a):
    return np.minimum(x + a * 128 * np.exp(-((x - 128) ** 2) / (2 *  ** 2)), 255)
```

```python
 = 70
a_values = [0.2, 0.5, 0.8, 1.0]
x = np.arange(0, 256)
```

```python
plt.figure(figsize=(10,6))
for a in a_values:
    y = vibrance_transform(x, a)
    plt.plot(x, y, label=f'a = {a}')

plt.plot(x, x, 'k--', label='Identity (no change)')
plt.title("Vibrance Transformation for Different 'a' Values")
plt.xlabel("Original Saturation (x)")
plt.ylabel("Transformed Saturation f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
a_values = [ 0.5, 0.6, 0.7, 0.8]

plt.figure(figsize=(10,6))
for a in a_values:
    y = vibrance_transform(x, a)
    plt.plot(x, y, label=f'a = {a}')

plt.plot(x, x, 'k--', label='Identity (no change)')
plt.title("Vibrance Transformation for Different 'a' Values")
plt.xlabel("Original Saturation (x)")
plt.ylabel("Transformed Saturation f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
a1 = 0.5
S_new1 = vibrance_transform(S, a1)

img_hsv_fixed1 = cv2.merge([H, S_new1, V])
img_rgb_fixed1 = cv2.cvtColor(img_hsv_fixed1.astype(np.uint8), cv2.COLOR_HSV2RGB)

a2 = 0.6
S_new2 = vibrance_transform(S, a2)

img_hsv_fixed2 = cv2.merge([H, S_new2, V])
img_rgb_fixed2 = cv2.cvtColor(img_hsv_fixed2.astype(np.uint8), cv2.COLOR_HSV2RGB)

a3 = 0.7
S_new3 = vibrance_transform(S, a3)

img_hsv_fixed3 = cv2.merge([H, S_new3, V])
img_rgb_fixed3 = cv2.cvtColor(img_hsv_fixed3.astype(np.uint8), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('original')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_rgb_fixed1)
plt.title('img_rgb_fixed a2 = 0.5')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(img_rgb_fixed2)
plt.title('img_rgb_fixed  a2 = 0.6')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(img_rgb_fixed3)
plt.title('img_rgb_fixed  a2 = 0.7')
plt.axis('off')
```
