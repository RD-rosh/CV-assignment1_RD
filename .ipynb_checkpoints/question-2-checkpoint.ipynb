```python
#gray matter
xp_gm = [0, 80, 150, 255]
fp_gm = [0, 120, 180, 255]

x = np.arange(256)
lookup_gm = np.interp(x, xp_gm, fp_gm).astype(np.uint8)

gm_img = cv2.LUT(img, lookup_gm)

#plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gm_img, cmap='gray')
plt.title("Gray matter img")
plt.axis('off')

plt.show()
```
